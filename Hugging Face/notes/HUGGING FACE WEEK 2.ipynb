1. Datasets library is an open-source library that provides efficient loading, preprocessing, and management of large datasets for machine learning and NLP tasks.

3.

| Topic                                | Syntax                                         | Example                                                                            | Detailed Description                                                                                                                          |
| ------------------------------------ | ---------------------------------------------- | ---------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------- |
| Loading Dataset from Hub             | `load_dataset("dataset_name")`                 | `dataset = load_dataset("imdb")`                                                   | Downloads dataset directly from Hugging Face Hub. Automatically handles splits, caching, and format conversion using Apache Arrow.            |
| Dataset Structure                    | `dataset["split_name"]`                        | `dataset["train"]`                                                                 | Datasets are stored as **DatasetDict** containing splits like train, test, validation. Each split behaves like a structured table.            |
| Viewing Dataset Samples              | `dataset["train"][index]`                      | `dataset["train"][0]`                                                              | Retrieves a single row as a dictionary. Useful for understanding fields like text, label, id and general data format.                         |
| Dataset Features                     | `dataset.features`                             | `print(dataset.features)`                                                          | Displays dataset schema including column names and data types such as string, integer, and ClassLabel. Helps understand label encoding.       |
| Dataset Size                         | `len(dataset["split"])`                        | `len(dataset["train"])`                                                            | Returns number of samples in a split. Important for estimating training time, batch planning, and memory requirements.                        |
| Loading Custom Files (CSV/JSON/Text) | `load_dataset(format, data_files="file")`      | `load_dataset("csv", data_files="data.csv")`                                       | Loads local datasets in common formats. Automatically parses columns and converts them into Dataset objects.                                  |
| Dataset Splitting                    | `.train_test_split(test_size=ratio)`           | `dataset.train_test_split(test_size=0.2)`                                          | Splits dataset into training and testing sets randomly based on ratio while preserving data distribution.                                     |
| Dataset Preprocessing                | `dataset.map(function)`                        | `dataset.map(preprocess)`                                                          | Applies preprocessing transformations to every row such as cleaning text, removing symbols, or normalizing case. Supports batched operations. |
| map() Transformation                 | `dataset.map(func)`                            | `dataset.map(lambda x: {"text": x["text"].lower()})`                               | Core function for dataset modification. Can add, remove, or modify columns efficiently with parallel processing support.                      |
| Removing Columns                     | `dataset.remove_columns(["col"])`              | `dataset.remove_columns(["id"])`                                                   | Deletes unnecessary features not required for training. Reduces dataset size and memory usage.                                                |
| Renaming Columns                     | `dataset.rename_column(old, new)`              | `dataset.rename_column("review","text")`                                           | Renames fields to standard names expected by transformer models (e.g., "text", "label") ensuring compatibility.                               |
| Selecting Rows                       | `dataset.select(range(n))`                     | `dataset.select(range(1000))`                                                      | Extracts a subset of rows for quick experiments, debugging, or small-scale testing.                                                           |
| Shuffling Dataset                    | `dataset.shuffle(seed=value)`                  | `dataset.shuffle(seed=42)`                                                         | Randomizes row order to prevent training bias and improve generalization. Seed ensures reproducibility.                                       |
| Filtering Dataset                    | `dataset.filter(function)`                     | `dataset.filter(lambda x: x["label"]==1)`                                          | Keeps only rows satisfying a condition. Useful for class balancing, noise removal, or domain filtering.                                       |
| Tokenizing Dataset                   | `dataset.map(tokenize_func)`                   | `dataset.map(tokenize_function)`                                                   | Applies tokenizer to text column converting sentences into input_ids and attention_mask required for transformer models.                      |
| Setting Format for PyTorch           | `dataset.set_format(type="torch", columns=[])` | `dataset.set_format(type="torch", columns=["input_ids","attention_mask","label"])` | Converts dataset outputs into PyTorch tensors enabling direct training with deep learning models.                                             |
