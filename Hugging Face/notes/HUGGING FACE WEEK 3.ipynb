# TOKENIZATION AND WORD EMBEDDINGS

| Point | Topic                         | Syntax                                                           | Example                                                               | Detailed Description                                                                                                                                                    | Objective                                              |
| ----- | ----------------------------- | ---------------------------------------------------------------- | --------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------ |
| 5     | AutoTokenizer Loading         | `AutoTokenizer.from_pretrained(model)`                           | `tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")`      | Automatically loads the correct tokenizer for a given pretrained model including vocabulary, merges, and special tokens. Ensures compatibility with model architecture. | Simplify tokenizer setup without manual configuration. |
| 6     | Tokenizing Single Sentence    | `tokenizer(text)`                                                | `tokenizer("I love NLP")`                                             | Converts text into token IDs and attention masks. Adds special tokens and handles padding/truncation if specified.                                                      | Prepare raw sentence for transformer input.            |
| 7     | Tokenizing Multiple Sentences | `tokenizer(list_of_texts, padding=True)`                         | `tokenizer(["I love NLP","Transformers are powerful"], padding=True)` | Batch tokenization for multiple inputs. Ensures equal sequence lengths via padding for efficient parallel processing.                                                   | Enable batch inference and training.                   |
| 8     | Tokenizer Output Fields       | `tokens.keys()`                                                  | `dict_keys(['input_ids','attention_mask'])`                           | Tokenizer returns structured dictionary containing numerical tensors needed for transformer models. May include token_type_ids for sentence pairs.                      | Understand model input components.                     |
| 9     | Decoding Tokens               | `tokenizer.decode(ids)`                                          | `tokenizer.decode([101, 1045, 2293, 17953, 2361, 102])`               | Converts token IDs back to readable text. Helps verify tokenization correctness and debug preprocessing issues.                                                         | Validate model inputs and outputs.                     |
| 10    | Special Tokens                | `tokenizer.cls_token`, `tokenizer.sep_token`                     | `[CLS] sentence [SEP]`                                                | Special tokens mark sequence boundaries and classification points. Different models use different tokens. Crucial for classification and QA tasks.                      | Provide structural signals to transformer models.      |
| 11    | Padding & Truncation          | `tokenizer(text, padding=True, truncation=True, max_length=128)` | `tokenizer(texts, padding="max_length", truncation=True)`             | Padding ensures equal sequence length; truncation limits overly long inputs. Necessary for fixed-size tensor processing.                                                | Maintain uniform batch shapes for GPU training.        |
| 12    | Return Tensors Format         | `tokenizer(text, return_tensors="pt")`                           | `tokenizer("Hello", return_tensors="pt")`                             | Converts tokenizer output directly into PyTorch tensors instead of Python lists, enabling direct model input.                                                           | Streamline preprocessing-to-model pipeline.            |
| 17    | Data Collator                 | `DataCollatorWithPadding(tokenizer)`                             | `collator = DataCollatorWithPadding(tokenizer)`                       | Dynamically pads sequences within each batch instead of entire dataset. Improves memory efficiency and training speed.                                                  | Optimize batching performance.                         |
| 18    | Dynamic vs Static Padding     | Static: pad to max_length  Dynamic: pad to longest in batch      | Dynamic padding example via data collator                             | Static padding wastes memory; dynamic padding adjusts to batch needs. Modern training pipelines prefer dynamic padding.                                                 | Reduce computation overhead.                           |
| 19    | Preparing Inputs for Model    | `model(**inputs)`                                                | `outputs = model(**tokens)`                                           | Tokenized outputs are unpacked and fed into transformer model. Inputs must include input_ids and attention_mask tensors.                                                | Run forward pass for predictions or training.          |
