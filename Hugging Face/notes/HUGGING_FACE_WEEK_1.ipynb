# ANSWERS

1. Hugging Face is an open-source AI platform that provides pretrained transformer models, datasets, and tools for building, fine-tuning, and deploying machine learning models, especially for Natural Language Processing (NLP) tasks.

2. Instead of training models from scratch, developers can download pretrained models and fine-tune them for specific tasks.

3. | Component        | Purpose                                  |
| ---------------- | ---------------------------------------- |
| Transformers     | Pretrained transformer models            |
| Datasets         | Efficient dataset loading and processing |
| Tokenizers       | Fast text tokenization                   |
| Hugging Face Hub | Repository for models and datasets       |
| Accelerate       | Distributed training                     |
| Evaluate         | Model evaluation metrics                 |
| Spaces           | Deploy ML apps                           |

4. The Hugging Face Model Hub is a centralized platform containing thousands of pretrained models and datasets.

   Developers can:

    > Download pretrained models

    > Upload trained models

    > Share datasets

    > Deploy applications

5. Pipelines provide a high-level API for performing inference with pretrained models. They automatically handle : Tokenization, Model loading, Preprocessing, Prediction and Postprocessing.

6. Common Hugging face task pipelines
   | Pipeline            | Purpose              |
| ------------------- | -------------------- |
| sentiment-analysis  | Detect sentiment     |
| text-generation     | Generate text        |
| fill-mask           | Predict masked words |
| question-answering  | Answer questions     |
| summarization       | Summarize text       |
| translation         | Translate languages  |
| text-classification | Classify text        |

7. Hugging Face provides Auto Classes that automatically select the correct model architecture. And its advantages are -

    > No need to manually select architecture

    > Automatically loads correct tokenizer and model
8. batch is NOT a single example but it is a dictionary of lists.
9. "truncation=True" means if the sentence is too long, cut it.
10. padding="max_length" means making  all sequences of same length.

1. Authentication is required to push models or datasets to the hub.

   Step 3 - Login using CLI

       huggingface-cli login

   Step 4 - Paste your access token.


2. Required Libraries

       pip install transformers
       pip install datasets
       pip install sentence-transformers
       pip install huggingface_hub
    
    Import Libraries

       from transformers import pipeline
       from transformers import AutoTokenizer, AutoModel
       from datasets import load_dataset

3. Using Pipeline for any task

        from transformers import pipeline

        pipeline("task_name")
  
   Sentiment Analysis Example -
      
        from transformers import pipeline

        classifier = pipeline("sentiment-analysis")

        classifier("I love machine learning")

        OUTPUT -

        [{'label': 'POSITIVE', 'score': 0.99}]
